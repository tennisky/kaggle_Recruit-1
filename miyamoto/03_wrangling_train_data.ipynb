```python
# libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import preprocessing
from datetime import datetime as dt
```

```python
tr=pd.read_csv('../dataset/train_data.csv')
```

```python
print(tr.shape)
tr.head()
```

```python
def group_mean(df,col):
    buf=df.groupby(['air_store_id',col]).mean()['visitors']
    buf=buf.reset_index()
    df1=buf.rename(columns={'visitors':'mean_'+col})
    return df1
```

```python
def group_max(df,col):
    buf=df.groupby(['air_store_id',col]).max()['visitors']
    buf=buf.reset_index()
    df1=buf.rename(columns={'visitors':'max_'+col})
    return df1
```

```python
def group_median(df,col):
    buf=df.groupby(['air_store_id',col]).median()['visitors']
    buf=buf.reset_index()
    df1=buf.rename(columns={'visitors':'median_'+col})
    return df1
```

```python
col=['year','month','day','day_of_week','working_day','holiday_flg','season']
```

```python
for i in col:
    gr=group_mean(tr,i)
    tr=pd.merge(tr,gr,on=['air_store_id',i],how='left')
```

```python
for i in col:
    gr=group_max(tr,i)
    tr=pd.merge(tr,gr,on=['air_store_id',i],how='left')
```

```python
for i in col:
    gr=group_median(tr,i)
    tr=pd.merge(tr,gr,on=['air_store_id',i],how='left')
```

```python
print(tr.shape)
tr.head()
```

```python
tr.set_index('store_id_date',drop=True,inplace=True)
tr.to_csv('../dataset/train_data1.csv')
```
