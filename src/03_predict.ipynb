```python
# libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import preprocessing
from datetime import datetime as dt
import pickle
import xgboost as xgb
```

```python
with open('../model/xgb1.pickle', mode='rb') as f:
    xgbst=pickle.load(f)
```

```python
df=pd.read_csv('../dataset/updated_test1.csv')
```

```python
df.set_index('id',drop=True,inplace=True)
```

```python
X = df.drop('visitors', axis=1).as_matrix()
y = df['visitors']
```

```python
y= xgbst.predict(X)
```

```python
submission=pd.DataFrame(columns=['id','visitors'])
submission['id']=df.index
submission['visitors']=y
```

```python
submission.set_index('id',drop=True,inplace=True)
submission.to_csv('../dataset/submission.csv')
```
